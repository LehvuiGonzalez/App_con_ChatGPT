Importar librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

importar *DataFrame*

```python
url =
"https://raw.githubusercontent.com/gabrielawad/programacion-para-ingenieria/\
refs/heads/main/archivos-datos/matplotlib/3d_plot_data.csv"


df = pd.read_csv(url, sep=',', decimal='.')
df
```

GrÃ¡fico de superficie (Surface plot)

```python
def plot_surface():
    """Genera un grÃ¡fico de superficie 3D utilizando los datos de las columnas
    'X_Surface', 'Y_Surface', y 'Z_Surface' del DataFrame.

    Args:
        None

    Returns:
        None: Muestra el grÃ¡fico sin retornar ningÃºn valor.
    """
    X = np.array(df['X_Surface'])
    Y = np.array(df['Y_Surface'])
    Z = np.array(df['Z_Surface'])

    # Suponiendo que los datos son una malla cuadrada
    n = int(np.sqrt(len(X)))
    X = X.reshape(n, n)
    Y = Y.reshape(n, n)
    Z = Z.reshape(n, n)

    # Crear la figura y los ejes 3D
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Crear el grÃ¡fico de superficie
    ax.plot_surface(X, Y, Z, cmap='viridis')

    # TÃ­tulo y etiquetas de los ejes
    ax.set_title('GrÃ¡fico de Superficie 3D')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Mostrar el grÃ¡fico
    plt.show()

# Llamada a la funciÃ³n para graficar la superficie
plot_surface()
```

GrÃ¡fico de dispersiÃ³n 3D (3D scatter plot)

```python
def plot_scatter():
    """Genera un grÃ¡fico de dispersiÃ³n 3D utilizando los datos de las columnas
    'Scatter_X', 'Scatter_Y' y 'Scatter_Z' del DataFrame.

    Args:
        None

    Returns:
        None: Muestra el grÃ¡fico sin retornar ningÃºn valor.
    """
    X = np.array(df['Scatter_X'])
    Y = np.array(df['Scatter_Y'])
    Z = np.array(df['Scatter_Z'])

    # Crear la figura y los ejes 3D
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Crear el grÃ¡fico de dispersiÃ³n 3D
    ax.scatter(X, Y, Z, c='r', marker='o')

    # TÃ­tulo y etiquetas de los ejes
    ax.set_title('GrÃ¡fico de DispersiÃ³n 3D')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Mostrar el grÃ¡fico
    plt.show()

# Llamada a la funciÃ³n para graficar la dispersiÃ³n 3D
plot_scatter()
```

GrÃ¡fico de malla (Wireframe plot)

```python
def plot_wireframe():
    """Genera un grÃ¡fico de malla 3D utilizando los datos de las columnas
    'X_Surface', 'Y_Surface' y 'Mesh_Z' del DataFrame.

    Args:
        None

    Returns:
        None: Muestra el grÃ¡fico sin retornar ningÃºn valor.
    """
    X = np.array(df['X_Surface'])
    Y = np.array(df['Y_Surface'])
    Z = np.array(df['Mesh_Z'])

    # Suponiendo que los datos forman una malla regular
    n = int(np.sqrt(len(X)))
    X = X.reshape(n, n)
    Y = Y.reshape(n, n)
    Z = Z.reshape(n, n)

    # Crear la figura y los ejes 3D
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Crear el grÃ¡fico de malla
    ax.plot_wireframe(X, Y, Z, color='black')

    # TÃ­tulo y etiquetas de los ejes
    ax.set_title('GrÃ¡fico de Malla 3D')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Mostrar el grÃ¡fico
    plt.show()

# Llamada a la funciÃ³n para graficar la malla 3D
plot_wireframe()
```

GrÃ¡fico de contorno 3D (3D contour plot)

```python
def plot_contour():
    """Genera un grÃ¡fico de contorno 3D utilizando los datos de las columnas
    'X_Surface', 'Y_Surface', y 'Z_Surface' del DataFrame.

    Args:
        None

    Returns:
        None: Muestra el grÃ¡fico sin retornar ningÃºn valor.
    """
    X = np.array(df['X_Surface'])
    Y = np.array(df['Y_Surface'])
    Z = np.array(df['Z_Surface'])

    # Suponiendo que los datos forman una malla regular
    n = int(np.sqrt(len(X)))
    X = X.reshape(n, n)
    Y = Y.reshape(n, n)
    Z = Z.reshape(n, n)

    # Crear la figura y los ejes 3D
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Crear el grÃ¡fico de contorno 3D
    ax.contour3D(X, Y, Z, 50, cmap='viridis')

    # TÃ­tulo y etiquetas de los ejes
    ax.set_title('GrÃ¡fico de Contorno 3D')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Mostrar el grÃ¡fico
    plt.show()

# Llamada a la funciÃ³n para graficar el contorno 3D
plot_contour()
```

GrÃ¡fico de barras 3D (3D bar plot)

```python
def plot_bar():
    """Genera un grÃ¡fico de barras 3D utilizando los datos de las columnas
    'Bar_X', 'Bar_Y', 'Bar_Height' del DataFrame.

    Args:
        None

    Returns:
        None: Muestra el grÃ¡fico sin retornar ningÃºn valor.
    """
    X = np.array(df['Bar_X'])
    Y = np.array(df['Bar_Y'])
    Z = np.zeros_like(X)
    height = np.array(df['Bar_Height'])

    # Crear la figura y los ejes 3D
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Crear el grÃ¡fico de barras 3D
    ax.bar3d(X, Y, Z, dx=1, dy=1, dz=height, color='b', zsort='average')

    # TÃ­tulo y etiquetas de los ejes
    ax.set_title('GrÃ¡fico de Barras 3D')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Height')

    # Mostrar el grÃ¡fico
    plt.show()

# Llamada a la funciÃ³n para graficar las barras 3D
plot_bar()
```
